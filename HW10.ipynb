## 5a

```python
import numpy as np
import matplotlib.pyplot as plt 
import random
```

```python
def markov_simulation(num):
    State = []
    x = random.randint(1, 3)
    for i in range(num):
        State.append(x)
        if (x == 1):
            p = random.random()
            if (p <= 0.3):
                x = 1
            else:
                x = random.randint(1, 3)
        if (x == 2):
            p = random.random()
            if (p <= 0.6):
                x = 1
            else:
                x = random.randint(1, 3)
        if (x == 3):
            p = random.random()
            if (p <= 0.9):
                x = 1
            else:
                x = random.randint(1, 3)
    return State
```

```python
N = [10, 20, 30, 40, 50, 100, 200]

for i in N:
    plt.figure(figsize=(8,6))
    plt.step(range(1,i+1), markov_simulation(i), label = str(i))
    plt.yticks([1, 2, 3])
    plt.legend()
    plt.xlabel("Steps")
    plt.ylabel("States")
    plt.title("Various states for each of the %d steps" %(i))
```

## 5b

```python
steps = np.arange(1000, 11000, 1000)

p1 = []
p2 = []
p3 = []
for i in steps:
    States = markov_simulation(i)
    p1.append(States.count(1))
    p2.append(States.count(2))
    p3.append(States.count(3))

plt.figure(figsize= (8,6));
plt.plot(steps, p1, label = "State 1");
plt.plot(steps, p2, label = "State 2");
plt.plot(steps, p3, label = "State 3");
plt.xticks(steps)
plt.ylabel("Count of States");
plt.xlabel("Steps");
plt.title("Plot of the proportion of each state for N steps");
plt.legend();
```
